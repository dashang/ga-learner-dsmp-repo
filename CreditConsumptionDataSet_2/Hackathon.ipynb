```python
def model_and_performance(model, model_name, train_x, train_y, test_x, test_y,y_actual):
    model.fit(train_x, train_y)
    train_pred = model.predict(train_x)
    test_pred = model.predict(test_x)
    actual_pred = model.predict(y_actual)
    
    print('MSE', mean_squared_error(train_y, train_pred), 'MAE', mean_absolute_error(train_y, train_pred), 
          'R2', r2_score(train_y, train_pred), 'RMSE', math.sqrt(mean_squared_error(train_y, train_pred)))
    
    print('MSE', mean_squared_error(test_y, test_pred), 'MAE', mean_absolute_error(test_y, test_pred), 
          'R2', r2_score(test_y, test_pred), math.sqrt(mean_squared_error(test_y, test_pred)))
    print('RMSLE Score: ',train_y,test_pred)
    pd.DataFrame(actual_pred, columns=['ID','cc_cons']).to_csv(''+model_name+'.csv')
```
